```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Define domains
domains = ["Handwritten", "Synthetic", "Typed", "All", "Existing Test"]

# Example WER matrix (fill in missing values with np.nan)
WER_data = pd.DataFrame([
    [1.267, 3.626, 4.218, 3.037, 11.362],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [9.295, 5.646, 0.426, 5.122, 107.393],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [1.712, 2.607, 1.734, 2.018, 1.165]
], index=domains, columns=domains)

# CER matrix
CER_data = pd.DataFrame([
    [2.032, 6.500, 4.746, 4.426, 3.883],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [6.192, 6.106, 0.329, 4.209, 23.610],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [3.712, 4.798, 4.487, 4.332, 1.264]
], index=domains, columns=domains)

# F1 Character matrix
F1_char_data = pd.DataFrame([
    [0.243, 0.121, 0.167, 0.177, 0.010],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [0.128, 0.077, 0.555, 0.253, 0.005],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [0.030, 0.019, 0.015, 0.022, 0.534]
], index=domains, columns=domains)

# F1 Word matrix
F1_word_data = pd.DataFrame([
    [0.138, 0.031, 0.015, 0.061, 0.000],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [0.004, 0.0003, 0.572, 0.192, 0.000],
    [np.nan, np.nan, np.nan, np.nan, np.nan],
    [0.00014, 0.000, 0.0004, 0.0002, 0.028]
], index=domains, columns=domains)

# Plotting function
def plot_heatmap(data, title, cmap="coolwarm", fmt=".2f"):
    plt.figure(figsize=(4, 4))
    sns.heatmap(data, annot=True, fmt=fmt, cmap=cmap, linewidths=.1, square=True, cbar_kws={'label': title})
    plt.title(title)
    plt.ylabel("Train Domain")
    plt.xlabel("Test Domain")
    plt.xticks(rotation=45)
    plt.yticks(rotation=0)
    plt.tight_layout()
    plt.show()

# Plot all heatmaps
plot_heatmap(WER_data, "WER")
plot_heatmap(CER_data, "CER")
plot_heatmap(F1_char_data, "F1 Score (Character)")
plot_heatmap(F1_word_data, "F1 Score (Word)")
```
